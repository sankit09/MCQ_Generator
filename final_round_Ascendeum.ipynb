# Q1: Identify the top drivers of CTR (Click-Through Rate)

```python
import pandas as pd



# Compute CTR per feature
def compute_ctr(df, groupby_col):
    summary = df.groupby(groupby_col).agg(
        impressions=('clicked', 'count'),
        clicks=('clicked', 'sum')
    ).reset_index()
    summary['CTR (%)'] = (summary['clicks'] / summary['impressions']) * 100
    return summary.sort_values(by='CTR (%)', ascending=False)

# By Device
ctr_device = compute_ctr(df, 'device_type').reindex()
# By Country
ctr_country = compute_ctr(df, 'country').reindex()
# By Publisher (platform)
ctr_publisher = compute_ctr(df, 'publisher_id').reindex()
# By Campaign
ctr_campaign = compute_ctr(df, 'campaign_id').reindex()
```
